```python
import pandas as pd
import numpy as np
from openpyxl import Workbook
from openpyxl.utils.dataframe import dataframe_to_rows
```

```python
df_transaction = pd.read_csv(r'C:\Users\Saurabh Mulgaonkar\Jupyter_Notebooks\IT_Vedant_Files\Projects\Final_Project\Python_Files\cleaned_data\transaction_cleaned.csv')
```

```python
# str to datetime
df_transaction['OrderDate'] = pd.to_datetime(df_transaction['OrderDate'], format="%Y-%m-%d")
```

```python
# correcting below to their absolute values
df_transaction['SaleValue'] = np.absolute(df_transaction['SaleValue'])
df_transaction['OrderedQuantity'] = np.absolute(df_transaction['OrderedQuantity'])
```

## Pivot Table

```python
#2019 data - Select 2019 data and create pivot table as per SaleValue
pt_2019 = df_transaction[(df_transaction['OrderDate']>=pd.to_datetime('2019-01-01')) & (df_transaction['OrderDate']<=pd.to_datetime('2019-12-31'))]
pt_1 = pd.pivot_table(pt_2019, values =['SaleValue'],index=['OrderDate'],fill_value=0, columns = ['MerchClassDescription'], aggfunc = {'SaleValue':'sum'}).apply(lambda x: x*100/sum(x))
```

```python
#2020 data
pt_2020 = df_transaction[(df_transaction['OrderDate']>=pd.to_datetime('2020-01-01')) & (df_transaction['OrderDate']<=pd.to_datetime('2020-12-31'))]
pt_2 = pd.pivot_table(pt_2020, values =['SaleValue'],index=['OrderDate'],fill_value=0, columns = ['MerchClassDescription'], aggfunc = {'SaleValue':'sum'}).apply(lambda x: x*100/sum(x))#.to_csv('pivot_table3.csv')
```

```python
#2021 data
pt_2021 = df_transaction[(df_transaction['OrderDate']>=pd.to_datetime('2021-01-01')) & (df_transaction['OrderDate']<=pd.to_datetime('2021-12-31'))]
pt_3 = pd.pivot_table(pt_2021, values =['SaleValue'],index=['OrderDate'],fill_value=0, columns = ['MerchClassDescription'], aggfunc = {'SaleValue':'sum'}).apply(lambda x: x*100/sum(x))#.to_csv('pivot_table3.csv')
```

```python
pt_festive = pd.concat([pt_1,pt_2,pt_3])
```

```python
pt_festive.replace(np.nan,0.0,inplace=True)
```
